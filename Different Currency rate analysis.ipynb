```python
from forex_python.converter import CurrencyRates
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
cr = CurrencyRates()
data = {'USD': [cr.get_rate('USD', 'EUR', date) for date in pd.date_range('2022-01-01', '2022-12-31').tolist()]}
df = pd.DataFrame(data)
df.plot()
plt.show()
```
